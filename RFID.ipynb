```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('RFID Dataset - Sheet1-2.csv')
```

```python
print(data.head())
print(data.info())
```

```python
type_counts = data['Type'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(type_counts.index, type_counts.values, color='skyblue')
plt.xlabel('Vehicle Type')
plt.ylabel('Count')
plt.title('Vehicle Type Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Time taken'], bins=20, color='lightcoral', edgecolor='black')
plt.xlabel('Time Taken')
plt.ylabel('Frequency')
plt.title('Time Taken Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Range'], data['Time taken'], c='green', alpha=0.5)
plt.xlabel('Range')
plt.ylabel('Time Taken')
plt.title('Range vs. Time Taken')
plt.show()
```
