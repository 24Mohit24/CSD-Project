```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('RFID Dataset - Sheet1-2.csv')
```

```python
print(data.head())
print(data.info())
```

```python
type_counts = data['Type'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(type_counts.index, type_counts.values, color='skyblue')
plt.xlabel('Vehicle Type')
plt.ylabel('Count')
plt.title('Vehicle Type Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Time taken'], bins=20, color='lightcoral', edgecolor='black')
plt.xlabel('Time Taken')
plt.ylabel('Frequency')
plt.title('Time Taken Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Range'], data['Time taken'], c='green', alpha=0.5)
plt.xlabel('Range')
plt.ylabel('Time Taken')
plt.title('Range vs. Time Taken')
plt.show()
```

```python
# Data Visualization Example 4: Box Plot
# Create a box plot to visualize the distribution of 'time taken' for each 'type' of vehicle
plt.figure(figsize=(10, 6))
data.boxplot(column='Time taken', by='Type', vert=False)
plt.title('Time Taken Distribution by Vehicle Type')
plt.xlabel('Time Taken')
plt.ylabel('Vehicle Type')
plt.show()
```

```python
from pandas.plotting import scatter_matrix
attributes = ['Range', 'Time taken']
scatter_matrix(data[attributes], figsize=(10, 6))
plt.suptitle('Scatter Matrix of Numeric Attributes', y=1.02)
plt.show()
```

```python
location_counts = data['Location'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(location_counts.index, location_counts.values, color='lightblue')
plt.xlabel('Location')
plt.ylabel('Count')
plt.title('Vehicle Count by Location')
plt.xticks(rotation=45)
plt.show()
```
